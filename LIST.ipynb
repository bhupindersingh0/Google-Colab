```python
import matplotlib.pyplot as plt

Customer_name = ["ram", "shaam", "ravi", "sahil", "abhi", "bhavya", "Preet"]
States = ["Bihar","Orissa", "UP", "West Bengal","Karnatak", "AP", "HP"]
Product = ["x", "y", "z", "a", "b", "c", "d"]
Amount = [100, 500, 450, 623, 399, 254, 542]
cost = [89, 452, 450, 605, 350, 250, 530]
profit = []



gender = ["Male", "Female", "Female", "Male", "Male", "Female", "Male"]
unique_state = []
unique_product = []
total_sales = 0
sales_state=[]
unique_gender = []

for i in range(len(Amount)):
  profit.append(Amount[i]-cost[i])
print(profit)


for i in Amount:
  total_sales = total_sales + i
print(total_sales)

# state wise sale
for i in States:
  if i not in unique_state:
    unique_state.append(i)
    sales_state.append(0)

for i in range(len(States)):
  index = unique_state.index(States[i])
  sales_state[index] = sales_state[index] + Amount[i]
print(unique_state)
print(sales_state)


plt.bar(unique_state,sales_state)
plt.xlabel(unique_state)
plt.ylabel(sales_state)
plt.show()

#Product
for i in Product:
  if i not in unique_product:
    unique_product.append(i)
    sales_state.append(0)
print(unique_product)

for i in range(len(Product)):
  index = unique_product.index(Product[i])
  sales_state[index] = sales_state[index] + Amount[i]
print(unique_product)
print(sales_state)


"""plt.bar(unique_product,sales_state)
plt.xlabel(unique_product)
plt.ylabel(sales_state)
plt.show()"""


for i in gender:
  if i not in unique_gender:
    unique_gender.append(i)
    sales_state.append(0)
print(unique_gender)




```
